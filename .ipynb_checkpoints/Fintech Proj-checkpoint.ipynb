```python
# creating main tkinter window/toplevel 
master = Tk() 
master.title("ROBOGAIN")
#master.geometry("1600x800")

rowCount=0

# this wil create a label widget 
l01 = Label(master, text = "Roboooooooo").grid(row = rowCount, column = 0, columnspan=2,sticky = W, pady = 2) 

l11 = Label(master, text = "Ticker 1:") 
l12 = Label(master, text = "Amount($):") 
l13 = Label(master, text = "Start Date:")

l21 = Label(master, text = "Ticker 2:") 
l22 = Label(master, text = "Amount($):") 
l23 = Label(master, text = "Start Date:")

l31 = Label(master, text = "Ticker 3:") 
l32 = Label(master, text = "Amount($):") 
l33 = Label(master, text = "Start Date:")

l41 = Label(master, text = "Ticker 4:") 
l42 = Label(master, text = "Amount($):") 
l43 = Label(master, text = "Start Date:")

l51 = Label(master, text = "Ticker 5:") 
l52 = Label(master, text = "Amount($):") 
l53 = Label(master, text = "Start Date:")

l61 = Label(master, text = "Ticker:") 
l62 = Label(master, text = "Amount($):") 
l63 = Label(master, text = "Start Date:")

# grid method to arrange labels in respective 
l11.grid(row = rowCount+1, column = 0, sticky = W, pady = 2) 
l12.grid(row = rowCount+1, column = 2, sticky = W, pady = 2) 
l13.grid(row = rowCount+1, column = 4, sticky = W, pady = 2) 

l21.grid(row = rowCount+2, column = 0, sticky = W, pady = 2) 
l22.grid(row = rowCount+2, column = 2, sticky = W, pady = 2) 
l23.grid(row = rowCount+2, column = 4, sticky = W, pady = 2) 

l31.grid(row = rowCount+3, column = 0, sticky = W, pady = 2) 
l32.grid(row = rowCount+3, column = 2, sticky = W, pady = 2) 
l33.grid(row = rowCount+3, column = 4, sticky = W, pady = 2) 

l41.grid(row = rowCount+4, column = 0, sticky = W, pady = 2) 
l42.grid(row = rowCount+4, column = 2, sticky = W, pady = 2) 
l43.grid(row = rowCount+4, column = 4, sticky = W, pady = 2) 

l51.grid(row = rowCount+5, column = 0, sticky = W, pady = 2) 
l52.grid(row = rowCount+5, column = 2, sticky = W, pady = 2) 
l53.grid(row = rowCount+5, column = 4, sticky = W, pady = 2) 

l61.grid(row = rowCount+6, column = 0, sticky = W, pady = 2) 
l62.grid(row = rowCount+6, column = 2, sticky = W, pady = 2) 
l63.grid(row = rowCount+6, column = 4, sticky = W, pady = 2) 


# return NGlobalTotal
def sumAmount():
    title3 = Label(master, text="Invested $")
    title3.grid(row = 8, column = 3,pady = 2)
    
    N1=float(e12.get())
    checkN1 = Label(master, text=N1)
    checkN1.grid(row = 10, column = 3,pady = 2)
    
    N2=float(e22.get())
    checkN2 = Label(master, text=N2)
    checkN2.grid(row = 11, column = 3,pady = 2)
    
    N3=float(e32.get())
    checkN3 = Label(master, text=N3)
    checkN3.grid(row = 12, column = 3,pady = 2)
    
    N4=float(e42.get())
    checkN4 = Label(master, text=N4)
    checkN4.grid(row = 13, column = 3,pady = 2)
    
    N5=float(e52.get())
    checkN5 = Label(master, text=N5)
    checkN5.grid(row = 14, column = 3,pady = 2)
    
    Ntotal=N1+N2+N3+N4+N5
    NglobalTotal=Ntotal
    checkT = Label(master, text=Ntotal)
    checkT.grid(row = 15, column = 3,pady = 2)
    return(NglobalTotal)


# return a list of weights
def weightsCalcAndPrint():
    title4 = Label(master, text="%")
    title4.grid(row = 8, column = 4,pady = 2)
    
    W1=float(e12.get())/sumAmount()
    W1deci = str(round(W1, 2))
    W1deciFloat=float(W1deci)
    checkW1 = Label(master, text=W1deciFloat)
    checkW1.grid(row = 10, column = 4,pady = 2)
    
    W2=float(e22.get())/sumAmount()
    W2deci = str(round(W2, 2))
    W2deciFloat=float(W2deci)
    checkW2 = Label(master, text=W2deciFloat)
    checkW2.grid(row = 11, column = 4,pady = 2)
    
    W3=float(e32.get())/sumAmount()
    W3deci = str(round(W3, 2))
    W3deciFloat=float(W3deci)
    checkW3 = Label(master, text=W3deciFloat)
    checkW3.grid(row = 12, column = 4,pady = 2)
    
    W4=float(e42.get())/sumAmount()
    W4deci = str(round(W4, 2))
    W4deciFloat=float(W4deci)
    checkW4 = Label(master, text=W4deciFloat)
    checkW4.grid(row = 13, column = 4,pady = 2)
    
    W5=float(e52.get())/sumAmount()
    W5deci = str(round(W5, 2))
    W5deciFloat=float(W5deci)
    checkW5 = Label(master, text=W5deciFloat)
    checkW5.grid(row = 14, column = 4,pady = 2)
    
    x = np.array([W1deciFloat,W2deciFloat,W3deciFloat,W4deciFloat,W5deciFloat])
    b=W1deciFloat
    c=W2deciFloat
    d=W3deciFloat
    e=W4deciFloat
    f=W5deciFloat
    #weights=getWeights(b,c,d,e,f)
    return(x)

def printListOfWeights():
    print(weightsCalcAndPrint())
    
def ListOfTickers():
    #class 'list'
    Tickers=[str(e11.get()),str(e21.get()),str(e31.get()),str(e41.get()),str(e51.get())]
    #print(Tickers)
    return(Tickers)

def ListOfStartDates():
    StartDates=[str(e13.get()),str(e23.get()),str(e33.get()),str(e43.get()),str(e53.get())]
    #print(StartDates)
    return(StartDates)

def FirstStartDates():
    x=e13.get()
    y="'"+e13.get()+"'"
    z=str(y)
    return(z)

def PortFunc():
    #Column 0
    title0 = Label(master, text="Ticker")
    title0.grid(row = 8, column = 0,pady = 2)
    
    tik1 = Label(master, text=e11.get()).grid(row = 10, column = 0,pady = 2)
    tik2 = Label(master, text=e21.get()).grid(row = 11, column = 0,pady = 2)
    tik3 = Label(master, text=e31.get()).grid(row = 12, column = 0,pady = 2)
    tik4 = Label(master, text=e41.get()).grid(row = 13, column = 0,pady = 2)
    tik5 = Label(master, text=e51.get()).grid(row = 14, column = 0,pady = 2)

    #Column 1
    #title1 = Label(master, text="Current Worth $")
    #title1.grid(row = 8, column = 1,pady = 2)
    
    #worth1 = Label(master, text="??").grid(row = 10, column = 1,pady = 2)  
    #worth2 = Label(master, text="??").grid(row = 11, column = 1,pady = 2)  
    #worth3 = Label(master, text="??").grid(row = 12, column = 1,pady = 2)  
    #worth4 = Label(master, text="??").grid(row = 13, column = 1,pady = 2)  
    #worth5 = Label(master, text="??").grid(row = 14, column = 1,pady = 2)  
    
    #Column 2
    title2 = Label(master, text="return %")
    title2.grid(row = 8, column = 2,pady = 2)
    
    r0 = Label(master, text="port: " +str(round(float(Average_return()),3))).grid(row = 10, column = 2,columnspan=2,pady = 2,sticky=W)  
    r1 = Label(master, text="bench: " +str(round(float(Average_ret_sp()),3))).grid(row = 9, column = 2,columnspan=2,pady = 2,sticky=W)  
    #r2 = Label(master, text="??").grid(row = 11, column = 2,pady = 2)  
    #r3 = Label(master, text="??").grid(row = 12, column = 2,pady = 2)  
    #r4 = Label(master, text="??").grid(row = 13, column = 2,pady = 2)  
    #r5 = Label(master, text="??").grid(row = 14, column = 2,pady = 2)  

def reportSum():
    tickers = ListOfTickers()
    SP_500 = ['^GSPC']
    weights = pd.Series(index = tickers, dtype = float)
    #weights[tickers]=0.2
    portweight = weightsCalcAndPrint()
    start_date = '2018-3-10'
    end_date = '2020-3-6'
    what_we_got = BK_get(tickers, start_date, end_date)
    SP_Close_Price = BK_get(SP_500, start_date, end_date)
    returns = what_we_got[tickers].pct_change()
    returns = returns.dropna()
    portfolio_return = returns.mul(portweight, axis=1).sum(axis=1)
    mu = returns.mean()
    sigma=mu.copy()
    for i in tickers:
        sigma[i]=stats.stdev(returns[i])
    what_we_got['portfolio']= portfolio_return
    what_we_got=what_we_got.dropna()
    cumulative_ret = (portfolio_return + 1).cumprod()
    #
    Average_return = np.mean(cumulative_ret)
    SP_500_ret = (SP_Close_Price.pct_change()).dropna()
    cumulative_SP_ret = (SP_500_ret).cumprod()
    Average_ret_SP_500 = np.mean(cumulative_SP_ret)
    standard_deviation = np.std(portfolio_return)
    mu_portfolio = np.mean(portfolio_return)
    #
    corr_df = returns.corr(method = 'pearson')
    avrg_ret = Average_return/100
    df1 = pd.DataFrame(portfolio_return)
    df2 = pd.DataFrame(SP_500_ret)
    #
    Excess_return= np.subtract(df1, df2)
    #
    tracking_error= np.std(Excess_return)
    risk_freerate = ['^TNX']
    series_code = 'DGS10'
    data_source = 'fred'
    #
    Yields = DataReader(series_code,data_source, start_date, end_date)
    #
    average_yield = np.mean(Yields) 
    avrg_yield = average_yield/100
    #
    sharpe_ratio = (mu_portfolio-0.0000556)/standard_deviation
    fig = plt.figure()
    ax1 = fig.add_axes([0.1,0.1,0.8,0.8])
    ax1.plot(cumulative_ret)
    ax1.set_xlabel('Date')
    ax1.set_ylabel("Cumulative Returns")
    ax1.set_title("Portfolio Cumulative Returns")
    #plt.show()
    return(round(sharpe_ratio,3))
    
def SharpeFunc():
    l1 = Label(master, text=str(reportSum())).grid(row = 18, column = 3,pady = 2)  
    
# graph in the main master.
def graphPort():
    window=Tk()
    window.title("Benchmark")
    #window.geometry('600x600')
    
    data1 = {'Country': ['US','CA','GER','UK','FR'],
             'GDP_Per_Capita': [45000,42000,52000,49000,47000]
            }
    df1 = DataFrame(data1,columns=['Country','GDP_Per_Capita'])


    data2 = {'Year': [1920,1930,1940,1950,1960,1970,1980,1990,2000,2010],
             'Unemployment_Rate': [9.8,12,8,7.2,6.9,7,6.5,6.2,5.5,6.3]
            }
    df2 = DataFrame(data2,columns=['Year','Unemployment_Rate'])


    data3 = {'Interest_Rate': [5,5.5,6,5.5,5.25,6.5,7,8,7.5,8.5],
             'Stock_Index_Price': [1500,1520,1525,1523,1515,1540,1545,1560,1555,1565]
            }  
    df3 = DataFrame(data3,columns=['Interest_Rate','Stock_Index_Price'])

    figure1 = plt.Figure(figsize=(6,5), dpi=100)
    ax1 = figure1.add_subplot(111)
    bar1 = FigureCanvasTkAgg(figure1, window)
    bar1.get_tk_widget().pack(side=tk.LEFT, fill=tk.BOTH)
    df1 = df1[['Country','GDP_Per_Capita']].groupby('Country').sum()
    df1.plot(kind='bar', legend=True, ax=ax1)
    ax1.set_title('Country Vs. GDP Per Capita')

    figure2 = plt.Figure(figsize=(5,4), dpi=100)
    ax2 = figure2.add_subplot(111)
    line2 = FigureCanvasTkAgg(figure2, window)
    line2.get_tk_widget().pack(side=tk.LEFT, fill=tk.BOTH)
    df2 = df2[['Year','Unemployment_Rate']].groupby('Year').sum()
    df2.plot(kind='line', legend=True, ax=ax2, color='r',marker='o', fontsize=10)
    ax2.set_title('Year Vs. Unemployment Rate')

    figure3 = plt.Figure(figsize=(5,4), dpi=100)
    ax3 = figure3.add_subplot(111)
    ax3.scatter(df3['Interest_Rate'],df3['Stock_Index_Price'], color = 'g')
    scatter3 = FigureCanvasTkAgg(figure3, window) 
    scatter3.get_tk_widget().pack(side=tk.LEFT, fill=tk.BOTH)
    ax3.legend(['Stock_Index_Price']) 
    ax3.set_xlabel('Interest Rate')
    ax3.set_title('Interest Rate Vs. Stock Index Price')

    
def getVar():
    # Initialize all the stuff you might want to change later
    tickers = ListOfTickers() # stock tickers go here
    # Need some weights for where we have our money
    weights = pd.Series(index = tickers, dtype = float)
    weights=weightsCalcAndPrint()
    start_date = '2018-3-10' # start day for lookback
    end_date = '2020-3-9' # end day for lookback

    # Initialize some Monte Carlo paramters
    monte_carlo_runs = 1000
    days_to_simulate = 10
    loss_cutoff = 0.97 # count any losses larger than 5% (or -5%)

    # Call that simple function we wrote above
    what_we_got = BK_get(tickers, start_date, end_date)
    # Compute returns from those Adjusted Closes
    returns = what_we_got[tickers].pct_change()
        # This is basically what bt.get returned, just the returns
    # Remove the NA from returns; we always get 1 fewer returns than data
    returns = returns.dropna() # pretty easy command

    # Calculate mu and sigma
    mu = returns.mean() # mu was very easy using .mean method
    # sigma was harder, as I couldn't do it in one line
    #   the statistics package didn't know what to do without the loop
    #   So I copied the mu data structure and filled it in a loop
    sigma=mu.copy() # copy the mu series (a pandas datatype)
    # Loop over tickers and fill sigma up with the calculated standard deviation
    for i in tickers: # python loops over tickers no problem
        sigma[i]=stats.stdev(returns[i]) # fill up sigma with stdev's
    # There is probably a cleaner way to do that 

    # Now we can start the Monte Carlo VaR loop

    compound_returns = sigma.copy()
    total_simulations = 0
    bad_simulations = 0
    for run_counter in range(0,monte_carlo_runs): # Loop over runs    
        for i in tickers: # loop over tickers, below is done once per ticker
            # Loop over simulated days:
            compounded_temp = 1
            for simulated_day_counter in range(0,days_to_simulate): # loop over days
                simulated_return = np.random.normal(mu[i],sigma[i],1)
                compounded_temp = compounded_temp * (simulated_return + 1)
            compound_returns[i]=compounded_temp # store compounded returns
        # Now see if those returns are bad by combining wi.th weights
        portfolio_return = compound_returns.dot(weights) # dot product
        if(portfolio_return<loss_cutoff):
            bad_simulations = bad_simulations + 1
        total_simulations = total_simulations + 1

    print("Your portfolio will lose",round((1-loss_cutoff)*100,3),"%",
          "over",days_to_simulate,"days", 
          bad_simulations/total_simulations, "of the time")
    xxVAR = "Your portfolio will lose " + str(round((1-loss_cutoff)*100,3)) + "% " + " over " + str(days_to_simulate) + " days " + str(bad_simulations/total_simulations) + " of the time."
    return(xxVAR)

def printVar():
    l1 = Label(master, text=str(getVar())).grid(row = 18, column = 4,columnspan=6,pady = 2)  
    
def Advise():
    #suggestWindow = Tk() 
    #suggestWindow.title("Tips")
    #master.geometry("1600x800")
    
    x=Average_return()
    y=float(Average_ret_sp())
    
    if x > y:
        print(" ")
        print("Your portfolio outperformed benchmark by: " + str(round(x-y,2)) + "%")
        #l1=Label(suggestWindow,text=x-y).grid(row=0,column=0)
    else:
        #l2=Label(suggestWindow,text=y-x).grid(row=0,column=1)
        print("Your portfolio underperformed benchmark by: " + str(round(y-x,2)) + "%")
        print("Diversify Portfolio.")
        print("Suggested Stocks: " + "GLD, UCO, BLV, UPRO, GSG")
   
    
def typeCheck():
    print(" ")
    print(type(Average_return()))
    print(type(float(Average_ret_sp())))
    

#--------------------------------------------------------------------------------------------------------------
def BK_get(tickers, s, e):
    stock_data = data.DataReader(tickers, data_source='yahoo', start = s, end = e)['Adj Close']
    return stock_data

def Average_return():
    tickers = ListOfTickers()
    SP_500 = ['^GSPC']
    weights = pd.Series(index = tickers, dtype = float)
    #weights[tickers]=0.5
    portweight = weightsCalcAndPrint()
    start_date = '2018-3-1'
    end_date = '2020-3-1'

    what_we_got = BK_get(tickers, start_date, end_date)
    SP_500_Close_Price = BK_get(SP_500, start_date, end_date)
    returns = what_we_got[tickers].pct_change()
    returns = returns.dropna()
    portfolio_return = returns.mul(portweight, axis=1).sum(axis=1)

    mu = returns.mean()
    sigma=mu.copy()
    for i in tickers:
        sigma[i]=stats.stdev(returns[i])

    what_we_got['portfolio']= portfolio_return
    what_we_got=what_we_got.dropna()

    cumulative_ret = (portfolio_return + 1).cumprod()

    Average_return = np.mean(cumulative_ret)*10

    sp_ret = (SP_500_Close_Price.pct_change()).dropna()
    cumulative_sp_ret = (sp_ret).cumprod()

    Average_ret_sp = np.mean(cumulative_sp_ret)*1000000

    corr_df = returns.corr(method = 'pearson')
    return(Average_return)

def Average_ret_sp():
    tickers = ListOfTickers()
    SP_500 = ['^GSPC']
    weights = pd.Series(index = tickers, dtype = float)
    #weights[tickers]=0.5
    portweight = weightsCalcAndPrint()
    start_date = '2018-3-1'
    end_date = '2020-3-1'

    what_we_got = BK_get(tickers, start_date, end_date)
    SP_500_Close_Price = BK_get(SP_500, start_date, end_date)
    returns = what_we_got[tickers].pct_change()
    returns = returns.dropna()
    portfolio_return = returns.mul(portweight, axis=1).sum(axis=1)

    mu = returns.mean()
    sigma=mu.copy()
    for i in tickers:
        sigma[i]=stats.stdev(returns[i])

    what_we_got['portfolio']= portfolio_return
    what_we_got=what_we_got.dropna()

    cumulative_ret = (portfolio_return + 1).cumprod()

    Average_return = np.mean(cumulative_ret)*10

    sp_ret = (SP_500_Close_Price.pct_change()).dropna()
    cumulative_sp_ret = (sp_ret).cumprod()

    Average_ret_sp = np.mean(cumulative_sp_ret)*1000000

    corr_df = returns.corr(method = 'pearson')
    print(corr_df)
    return(Average_ret_sp)
#--------------------------------------------------------------------------------------------------------------

# entry widgets, used to take entry from user 
e11 = Entry(master) 
e12 = Entry(master)
e13 = Entry(master,state=DISABLED) 

e21 = Entry(master) 
e22 = Entry(master)
e23 = Entry(master,state=DISABLED) 

e31 = Entry(master) 
e32 = Entry(master)
e33 = Entry(master,state=DISABLED) 

e41 = Entry(master) 
e42 = Entry(master)
e43 = Entry(master,state=DISABLED) 

e51 = Entry(master) 
e52 = Entry(master)
e53 = Entry(master,state=DISABLED) 

e61 = Entry(master,foreground ="pink") 
e61.insert(END,"Example: FB")
e62 = Entry(master,foreground ='pink')
e62.insert(END,"Example: 50000")
e63 = Entry(master,foreground ='pink') 
e63.insert(END,"Example: 2018-12-31")

# this will arrange entry widgets 
e11.grid(row = rowCount+1, column = 1, pady = 2) 
e12.grid(row = rowCount+1, column = 3, pady = 2) 
e13.grid(row = rowCount+1, column = 5, pady = 2) 

e21.grid(row = rowCount+2, column = 1, pady = 2) 
e22.grid(row = rowCount+2, column = 3, pady = 2) 
e23.grid(row = rowCount+2, column = 5, pady = 2) 

e31.grid(row = rowCount+3, column = 1, pady = 2) 
e32.grid(row = rowCount+3, column = 3, pady = 2) 
e33.grid(row = rowCount+3, column = 5, pady = 2) 

e41.grid(row = rowCount+4, column = 1, pady = 2) 
e42.grid(row = rowCount+4, column = 3, pady = 2) 
e43.grid(row = rowCount+4, column = 5, pady = 2) 

e51.grid(row = rowCount+5, column = 1, pady = 2) 
e52.grid(row = rowCount+5, column = 3, pady = 2) 
e53.grid(row = rowCount+5, column = 5, pady = 2) 

e61.grid(row = rowCount+6, column = 1, pady = 2) 
e62.grid(row = rowCount+6, column = 3, pady = 2) 
e63.grid(row = rowCount+6, column = 5, pady = 2) 

b0 = Button(master, text="Portfolio Return", width=12, command=PortFunc)
b0.grid(row = 7, column = 0, columnspan=5,pady=20)

b1 = Button(master, text="Total Amount", width=12,command = sumAmount)
b1.grid(row = 7, column = 1, columnspan=5)

b2 = Button(master, text="Weights", width=12,command = weightsCalcAndPrint)
b2.grid(row = 7, column = 2, columnspan=5)

#graphPort
Button(master,text="Port Summary", width=12,command=Advise).grid(row = 16, column =0, columnspan=5,pady=20)
Button(master,text="Sharpe", width=12,command=SharpeFunc).grid(row = 16, column = 1,columnspan=5)
Button(master,text="VAR", width=12,command=printVar).grid(row = 16, column = 2,columnspan=5)
    
#current rows used: 0-17
    
def close_window():
    master.destroy()
    exit()
Button(master,text="Exit",command=close_window).grid(row = 8, column = 6)      
Label(master,text="End date = today: " + str(today),state=DISABLED,relief="solid").grid(row = 8, column = 5)  


                         
mainloop()
```
